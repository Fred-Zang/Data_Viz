## 1. Pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Graphiques simples

```python
X = np.linspace(0, 2, 10)
y = X**2

plt.plot(X, y, ls='dashed')  # dashed = '--'
plt.show()
plt.subplot
```

```python
plt.scatter(X, y)
plt.show()
```

### Styles Graphiques
Il existe beaucoup de styles a ajouter aux graphiques. Voici les plus importants a retenir :
- **c** : couleur de la ligne
- **lw** : epaisseur de la ligne (pour les graphiques plot)
- **ls** : style de la ligne (pour les graphiques plot)
- **size** : taille du point (pour les graphiques scatter)
- **marker** : style de points (pour les graphiques scatter)
- **alpha** : transparence du graphique



```python
plt.plot(X, y, c='red', lw=4, ls='--', alpha=0.8)
plt.show()
```

## 2. Cycle de vie d'une figure
Pour crer des figures proprements, on doit suivre le cycle de vie suivant :
1. **plt.figure**(figsize=())
2. **plt.plot()**
3. Extras (titre, axes, legendes)
4. **plt.show()**

```python
X = np.linspace(0, 2, 10)

plt.figure() # Cration d'une figure - crire ceci pour chaque figure  afficher
plt.plot(X, X**2, label='quadratique') # premiere courbe avec son label ( pour legend  la fin)
plt.plot(X, X**3, label='cubique') # deuxieme courbe avec son label
# Extra information
plt.title('figure 1') # titre
plt.xlabel('axe x') # label de l'axe x
plt.ylabel('axe y') # label de l'axe y
plt.legend() #   encadr legend() en dernier avec les dfinitions des labels dans les plot avant

plt.savefig('figure.png') # sauvegarde la figure dans le repertoire de travail
plt.show() # affiche la figure
```

## 3. Subplots
Les subplots sont un autre lments a ajouter pour crer plusieurs graphiques sur une meme figure

```python
plt.figure(figsize=(12, 8))  # taille de la figure 12 inch en largeur et 8 inch en hauteur
x = np.linspace(-10, 10, 30)  # dfinir une ligne de 50 points x sur l'intervalle -10 ; +10
plt.subplot(2, 1, 1)        # 1er suplot dfinit
plt.plot(x, x**2, c='red')
plt.subplot(2, 1, 2)       # 2eme subplot dfinit
plt.plot(x, x**3, c='blue')    # PS: le choix des valeurs des axes se fait automatiquement
```

## 4. Mthode oriente objet

```python
fig, ax = plt.subplots(2, 1, sharex=True) # partage le meme axe pour les subplots
x = np.linspace(-10, 10, 50)  # dfinir une ligne de 50 points x sur l'intervalle -10 ; +10
y = x**2
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```

## 5. Exercice et Solution
Crez une fonction "graphique" qui permet de tracer sur une seule et meme figure une srie de graphiques issue d'un dictionnaire contenant plusieurs datasets :

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)} # on cr un dictionaire de 4 expriences alatoires
# randn = chantillons de la distribution normale standard (alias gaussienne) (moyenne 0 et variance 1).
# randn = liste [ ] ici car 1 seul nombre (=100) dans randn, si randn(100, 3) alors matrice 100x3
# dataset = dictionnaire de donnes = {"experience0":[donnes0], ...}
# dataset  # -> {'experience0': array([-3.88450747e-01, -6.04284968e-01, -6.50146503e-01,  5.20016446e-01, ......
       
```

```python
# SOLUTION
def graphique(data):    #  la base on a un data = dictionnaire {exprience0:[liste data 1]....}
    n = len(data) #  -> n = 4  longeur de notre dataset
    
    plt.figure(figsize=(12, 20))  # on cr une figure tape 1
   
    # cration de la boucle d'affichage des 4 graph dans le mme subplot ( fct zip relie des listes ou chaine en 1 seule liste
    for k, i in zip(data.keys(), range(1, n+1)):  # i entre 1 et n=4, on veut le nombre d'expriences ET le contenu du dictionnaire cl k
        plt.subplot(n, 1, i)   # grille de graphs de n lignes, 1 colonne, chaque subplot i tout  tour
        plt.xticks(color='red')  # on dfinit la couleur des axes x et u
        plt.yticks(color='yellow')
        plt.plot(data[k])   # on indique sur quel tracer ici le graph data[k]
        plt.title(k)  # titre sur la clk

        
    plt.show()
```

```python
graphique(dataset)
```
